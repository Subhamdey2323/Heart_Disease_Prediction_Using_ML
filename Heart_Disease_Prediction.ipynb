```python
heart_data.drop(['education'], inplace = True, axis = 1)
heart_data.rename(columns ={'male':'Sex'}, inplace = True)
```

```python
X = np.asarray(heart_data[['age', 'Sex', 'cigsPerDay',
                           'totChol', 'sysBP', 'glucose','diaBP','BMI','heartRate']])
y = np.asarray(heart_data['TenYearCHD'])

# normalization of the dataset
from sklearn import preprocessing

X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Training Accuracy is : ",accuracy_score(y_train, model.predict(X_train)))
print("Testing Accuracy is : ",accuracy_score(y_test, model.predict(X_test)))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
conf_matrix = pd.DataFrame(data = cm,
                           columns = ['Predicted:0', 'Predicted:1'],
                           index =['Actual:0', 'Actual:1'])

plt.figure(figsize = (12, 5))
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens")

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model1 = DecisionTreeClassifier(max_depth=6)
model1.fit(X_train, y_train)

# testing the model
print("Training Accuracy is : ",accuracy_score(y_train, model1.predict(X_train)))
print("Testing Accuracy is : ",accuracy_score(y_test, model1.predict(X_test)))
```

```python
# Confusion matrix of Results from Decision Tree classification
from sklearn import metrics

cm1 = metrics.confusion_matrix(y_test, model1.predict(X_test))

cm_display1 = metrics.ConfusionMatrixDisplay(confusion_matrix=cm1,display_labels=[False, True])

cm_display1.plot()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier()
model2.fit(X_train, y_train)

# testing the model
print("Training Accuracy is : ",accuracy_score(y_train, model2.predict(X_train)))
print("Testing Accuracy is : ",accuracy_score(y_test, model2.predict(X_test)))
```
